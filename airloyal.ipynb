#Clear Missing Data

```python
og_data = og_data.dropna().reset_index(drop = True)
og_data = og_data.drop(labels=["id"], axis=1)
og_data = og_data.drop(labels=["Arrival Delay in Minutes"], axis=1)
og_data = og_data.drop(labels=["Departure Delay in Minutes"], axis=1)
```

```python
df = og_data.copy()
df["Customer Type"] = df["Customer Type"].replace(["disloyal Customer", "Loyal Customer"],[0 , 1] )
df["Gender"] = df["Gender"].replace(["Male", "Female"],[0 , 1] )
df["satisfaction"] = df["satisfaction"].replace(["neutral or dissatisfied", "satisfied"],[0 , 1] )
df["Type of Travel"] = df["Type of Travel"].replace(["Business travel", "Personal Travel"],[0 , 1] )

```

```python
df
```

#Add new column

```python
df['Business'] = df['Class']
df['Eco Plus'] = df['Class']
df['Eco'] = df['Class']
df["Business"] = df["Business"].replace(["Business", "Eco Plus" , "Eco"],[1 , 0 , 0] )
df["Eco Plus"] = df["Eco Plus"].replace(["Business", "Eco Plus" , "Eco"],[0 , 1 , 0] )
df["Eco"] = df["Eco"].replace(["Business", "Eco Plus" , "Eco"],[0 , 0 , 1] )
df = df.drop(labels=["Class"], axis=1)
df
```

#outliner clear


```python
for column in df.columns:
  if (df[column].dtypes == 'int64' and column != 'Eco Plus'):
   df = df[np.abs(df[column]-
           df[column].mean()) <=
            (3*df[column].std())]
df
```
